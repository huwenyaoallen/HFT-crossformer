```python
preds = pd.read_csv('./results/Crossformer_Stock30min_il96_ol-1_sl8_win2_fa10_dm256_nh4_el3_itr0/pred.csv')
trues = pd.read_csv('./results/Crossformer_Stock30min_il96_ol-1_sl8_win2_fa10_dm256_nh4_el3_itr0/true.csv')
```

```python
#compute the pearson correlation coefficient row-wise
from scipy.stats import spearmanr
ICs = []
for i in range(len(preds)):
    #drop the first column, which is the timestamp
    pred = preds.iloc[i, 1:]
    true = trues.iloc[i, 1:]

    #check if there are any NaN values, if so, drop them
    if pred.isna().sum() > 0:
        pred = pred.dropna()
        true = true.dropna()
    
    #compute the correlation coefficient
    rankic = spearmanr(pred, true)[0]
    ICs.append(rankic)
ICs = np.array(ICs)
```

```python
ICs.mean()
```
